## Definir função gráfica



Nesta seção, definimos a função `make_graph`. **Você não precisa saber como a função funciona, você deve se importar apenas com as entradas. Ela pega um dataframe com dados de ações (o dataframe deve conter as colunas Date e Close), um dataframe com dados de receita (o dataframe deve conter as colunas Date e Revenue) e o nome da ação.**


````python
def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing = .3)
    stock_data_specific = stock_data[stock_data.Date <= '2021--06-14']
    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data_specific.Date), y=stock_data_specific.Close.astype("float"), name="Share Price"), row=1, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data_specific.Date), y=revenue_data_specific.Revenue.astype("float"), name="Revenue"), row=2, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    fig.update_layout(showlegend=False,
    height=900,
    title=stock,
    xaxis_rangeslider_visible=True)
    fig.show()
````

## Pergunta 2: Use Web Scraping para extrair dados de receita da Tesla


Use a biblioteca `requests` para baixar a página da web https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm Salve o texto da resposta como uma variável chamada `html_data`.


````python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'
html_data = requests.get(url)
````

Analise os dados html usando `beautiful_soup` usando o parser, ou seja, `html5lib` ou `html.parser`. Certifique-se de usar `html_data` com o parâmetro content como segue `html_data.content` .


````python
beautiful_soup = BeautifulSoup(html_data.content, "html.parser")
````

Usando `BeautifulSoup` ou a função `read_html` extraia a tabela com `Tesla Revenue` e armazene-a em um dataframe chamado `tesla_revenue`. O dataframe deve ter as colunas `Date` e `Revenue`.


<details><summary>Instruções passo a passo</summary>

```

Aqui estão as instruções passo a passo:

1. Encontre todas as tabelas: comece pesquisando todas as tabelas HTML em uma página da web usando `soup.find_all('table')`.
2. Identifique a tabela relevante: então faça um loop em cada tabela. Se uma tabela contiver o texto "Tesla Quarterly Revenue", selecione essa tabela.
3. Inicialize um DataFrame: crie um Pandas DataFrame vazio chamado `tesla_revenue` com as colunas "Date" e "Revenue".
4. Faça um loop nas linhas: para cada linha na tabela relevante, extraia os dados da primeira e segunda colunas (data e receita).
5. Limpe os dados da receita: remova os cifrões e as vírgulas do valor da receita.
6. Adicione linhas ao DataFrame: crie uma nova linha no DataFrame com a data extraída e os valores de receita limpos.
7. Repita para todas as linhas: continue esse processo para todas as linhas na tabela.

```
</details>


<details><summary>Clique aqui se precisar de ajuda para localizar a tabela</summary>

```
    
Abaixo está o código para isolar a tabela, agora você precisará fazer um loop pelas linhas e colunas como no laboratório anterior

soup.find_all("tbody")[1]

Se você quiser usar a função read_html, a tabela está localizada no índice 1

Estamos focando na receita trimestral no laboratório.
> Nota: Em vez de usar o método obsoleto pd.append(), considere usar pd.concat([df, pd.DataFrame], ignore_index=True).
```

</details>


````python
# Encontre todas as tabelas: comece pesquisando todas as tabelas HTML em uma 
# página da web usando `soup.find_all('table')`.
tables = beautiful_soup.find_all('table') 

# Identifique a tabela relevante: então faça um loop em cada tabela. 
# Se uma tabela contiver o texto "Tesla Quarterly Revenue", selecione tabela.
for i, table in enumerate(tables): 
    if ("Tesla Quarterly Revenue" in str(table)): #
        table_index = i

# Inicialize um DataFrame: crie um Pandas DataFrame vazio chamado `tesla_revenue`
#  com as colunas "Date" e "Revenue".
tesla_revenue = pd.DataFrame(columns=['Date', 'Revenue']) 

# Faça um loop nas linhas: para cada linha na tabela relevante, extraia os dados 
# da primeira e segunda colunas (data e receita).
for row in tables[table_index].tbody.find_all('tr'): 
    col = row.find_all('td')
    if (col != []):
        date = col[0].text.strip()
        revenue = col[1].text.strip()
        row_data = pd.DataFrame({'Date': [date], 'Revenue': [revenue]})
        tesla_revenue = pd.concat([tesla_revenue, row_data], ignore_index=True)

tesla_revenue
````

Execute a seguinte linha para remover a vírgula e o cifrão da coluna `Receita`.


````python
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"", regex=True)
````

Execute as seguintes linhas para remover strings nulas ou vazias na coluna Receita.


````python
tesla_revenue.dropna(inplace=True)

tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
````

Exiba as últimas 5 linhas do dataframe `tesla_revenue` usando a função `tail`. Tire uma captura de tela dos resultados.


````python
tesla_revenue.tail() # retorna as 5 linhas inferiores por padrão
# se vc quiser um retorne de quntidade diferente e só colocar na função .tail(2)
````

## Question 4: Use Webscraping to Extract GME Revenue Data

Analise os dados html usando `beautiful_soup` usando o analisador, ou seja, `html5lib` ou `html.parser`.


````python
soup = BeautifulSoup(html_data_2.content, "html.parser")
````

Usando `BeautifulSoup` ou a função `read_html`, extraia a tabela com `GameStop Revenue` e armazene-a em um dataframe chamado `gme_revenue`. O dataframe deve ter as colunas `Date` e `Revenue`. Certifique-se de que a vírgula e o cifrão sejam removidos da coluna `Revenue`.


````python
table1 = soup.find_all('tbody')

gme_revenue = pd.DataFrame(columns=['Date', 'Revenue'])

# Verifica isso aqui minha filha 
for row in table1[1]('tr'): 
    col = row.find_all('td')
    if (col != []):
        date = col[0].text.strip()
        revenue = col[1].text.strip()
        row_data = pd.DataFrame({'Date': [date], 'Revenue': [revenue]})
        gme_revenue = pd.concat([gme_revenue, row_data], ignore_index=True)

print(gme_revenue)

gme_revenue.dropna(inplace=True)
gme_revenue["Revenue"] = gme_revenue['Revenue'].str.replace(',|\$',"", regex=True)

print(gme_revenue)
````
